```python
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].median())
df['Pregnancies'] = df['Pregnancies'].replace(0, df['Pregnancies'].median())
df['Glucose'] = df['Glucose'].replace(0, df['Glucose'].mean())
df['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'].mean())
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].median())
df['BMI'] = df['BMI'].replace(0, df['BMI'].mean())
df['DiabetesPedigreeFunction'] = df['DiabetesPedigreeFunction'].replace(0, df['DiabetesPedigreeFunction'].median())
df['Age'] = df['Age'].replace(0, df['Age'].median())
```

```python
fig, axs = plt.subplots(9,1,figsize=(10,50))
for i in range(len(df.columns)):
  sns.distplot(df[df.columns[i]], ax=axs[i])
  axs[i].set_title(a[i])
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
plt.figure(figsize=(12,7))

sns.boxplot(data=x)
```

```python
for i in x.columns:
  Q1 = x[i].quantile(0.25)
  Q3 = x[i].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  mask = (x[i] >= lower_bound) & (x[i] <= upper_bound)
```

```python
x_new=x[mask]
y_new=y[mask]
```

```python
y.shape
```

```python
y_new.shape
```

```python
plt.figure(figsize=(12,7))

sns.boxplot(data=x_new)
```

```python
x_1=x_new[(x_new['Insulin'])<(x_new['Insulin'].quantile(.98))]
y_1=y_new[(x_new['Insulin'])<(x_new['Insulin'].quantile(.98))]
```

```python
y_1
```

```python
plt.figure(figsize=(12,7))

sns.boxplot(data=x_1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x_1)
```

```python
x_norm=pd.DataFrame(x_scale, columns=x_new.columns)
```

```python

x_norm
```

```python
plt.figure(figsize=(12,7))

sns.boxplot(data=x_scale)
```

```python
y_1.value_counts()
```

```python
x_norm.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_norm, y_1, test_size=0.3, random_state=42)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_1, y_train_1 = smote.fit_resample(x_train, y_train)
```

```python
y_train_1.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_1, y_train_1)
```

```python
p1=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, p1)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, p1))
```

```python
import pickle
pickle.dump(model, open("diabetic_classification_model.pkl", "wb"))
```
